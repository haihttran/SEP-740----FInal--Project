```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
yolo11s_coco_df = pd.read_csv(r'yolo11s_coco.csv')
yolo11n_coco_df = pd.read_csv(r'yolo11n_coco.csv')
yolo11n_coco_distilled_df = pd.read_csv(r'yolo11n_coco_distilled.csv')
```

```python
display(yolo11s_coco_df)
```

```python
display(yolo11n_coco_df)
```

```python
display(yolo11n_coco_distilled_df)
```

```python
plt.figure(figsize =(5,5))
plt.plot(yolo11s_coco_df['time'],label='yolo11s')
plt.plot(yolo11n_coco_df['time'],label='yolo11n')
plt.plot(yolo11n_coco_distilled_df['time'],label='yolo11n distilled')
plt.legend()
plt.title('Training time (s)\n', 
                fontsize = 14, fontweight ='bold')
```

```python
plt.figure(figsize =(5,5))
plt.plot(yolo11s_coco_df['train/box_loss'],label='yolo11s')
plt.plot(yolo11n_coco_df['train/box_loss'],label='yolo11n')
plt.plot(yolo11n_coco_distilled_df['train/box_loss'],label='yolo11n distilled')
plt.legend()
plt.title('Training box loss\n', 
                fontsize = 14, fontweight ='bold')
```

```python
plt.figure(figsize =(5,5))
plt.plot(yolo11s_coco_df['train/cls_loss'],label='yolo11s')
plt.plot(yolo11n_coco_df['train/cls_loss'],label='yolo11n')
plt.plot(yolo11n_coco_distilled_df['train/cls_loss'],label='yolo11n distilled')
plt.legend()
plt.title('Training class loss\n', 
                fontsize = 14, fontweight ='bold')
```

```python
plt.figure(figsize =(5,5))
plt.plot(yolo11s_coco_df['metrics/precision(B)'],label='yolo11s')
plt.plot(yolo11n_coco_df['metrics/precision(B)'],label='yolo11n')
plt.plot(yolo11n_coco_distilled_df['metrics/precision(B)'],label='yolo11n distilled')
plt.legend()
plt.title('Precision\n', 
                fontsize = 14, fontweight ='bold')
```

```python
plt.figure(figsize =(5,5))
plt.plot(yolo11s_coco_df['metrics/mAP50(B)'],label='yolo11s')
plt.plot(yolo11n_coco_df['metrics/mAP50(B)'],label='yolo11n')
plt.plot(yolo11n_coco_distilled_df['metrics/mAP50(B)'],label='yolo11n distilled')
plt.legend()
plt.title('mean Average Precision 50 (mAP50)\n', 
                fontsize = 14, fontweight ='bold')
```

```python
plt.figure(figsize =(5,5))
plt.plot(yolo11s_coco_df['metrics/mAP50-95(B)'],label='yolo11s')
plt.plot(yolo11n_coco_df['metrics/mAP50-95(B)'],label='yolo11n')
plt.plot(yolo11n_coco_distilled_df['metrics/mAP50-95(B)'],label='yolo11n distilled')
plt.legend()
plt.title('mean Average Precision 50-95 (mAP50-95)\n', 
                fontsize = 14, fontweight ='bold')
```

```python
figure, axis = plt.subplots(3,2, figsize=(10,15))


axis[1, 0].plot(yolo11s_coco_df['metrics/mAP50(B)'],label='yolo11s')
axis[1, 0].plot(yolo11n_coco_df['metrics/mAP50(B)'],label='yolo11n')
axis[1, 0].plot(yolo11n_coco_distilled_df['metrics/mAP50(B)'],label='yolo11n distilled')
axis[1, 0].legend()
axis[1, 0].set_title("mean Average Precision 50 (mAP50)")
axis[1, 0].set_xlabel("epoch")


axis[1, 1].plot(yolo11s_coco_df['metrics/mAP50-95(B)'],label='yolo11s')
axis[1, 1].plot(yolo11n_coco_df['metrics/mAP50-95(B)'],label='yolo11n')
axis[1, 1].plot(yolo11n_coco_distilled_df['metrics/mAP50-95(B)'],label='yolo11n distilled')
axis[1, 1].legend()
axis[1, 1].set_title("mean Average Precision 50-95 (mAP50-95)")
axis[1, 1].set_xlabel("epoch")

axis[0, 0].plot(yolo11s_coco_df['train/box_loss'],label='yolo11s')
axis[0, 0].plot(yolo11n_coco_df['train/box_loss'],label='yolo11n')
axis[0, 0].plot(yolo11n_coco_distilled_df['train/box_loss'],label='yolo11n distilled')
axis[0, 0].legend()
axis[0, 0].set_title("Train Box Loss")
axis[0, 0].set_xlabel("epoch")

axis[0, 1].plot(yolo11s_coco_df['train/cls_loss'],label='yolo11s')
axis[0, 1].plot(yolo11n_coco_df['train/cls_loss'],label='yolo11n')
axis[0, 1].plot(yolo11n_coco_distilled_df['train/cls_loss'],label='yolo11n distilled')
axis[0, 1].legend()
axis[0, 1].set_title("Train Class Loss")
axis[0, 1].set_xlabel("epoch")

axis[2, 0].plot(yolo11s_coco_df['metrics/precision(B)'],label='yolo11s')
axis[2, 0].plot(yolo11n_coco_df['metrics/precision(B)'],label='yolo11n')
axis[2, 0].plot(yolo11n_coco_distilled_df['metrics/precision(B)'],label='yolo11n distilled')
axis[2, 0].legend()
axis[2, 0].set_title("Precision")
axis[2, 0].set_xlabel("epoch")

axis[2, 1].plot(yolo11s_coco_df['metrics/recall(B)'],label='yolo11s')
axis[2, 1].plot(yolo11n_coco_df['metrics/recall(B)'],label='yolo11n')
axis[2, 1].plot(yolo11n_coco_distilled_df['metrics/recall(B)'],label='yolo11n distilled')
axis[2, 1].legend()
axis[2, 1].set_title("Recall")
axis[2, 1].set_xlabel("epoch")
```
